```python
import numpy as np
import pandas as pd
import os
```

```python
os.chdir('D:\\Users\Dell\Desktop\Data Science\Datasets')
df = pd.read_csv('carprices.csv')
df
```

```python
dum = pd.get_dummies(df['Car Model'])
dum
```

```python
merged = pd.concat([df,dum],axis=1)
merged
```

```python
final = merged.drop(['Car Model','Audi A5'],1)               #to avoid dummy variable trap
final
```

```python
x=final.drop('Sell Price($)',1)
x
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)  
# an additional parameter 'random_state=10' will always generate same training set everytime
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.predict([[45000,4,0,1],[86000,7,1,0]])
```

```python
model.score(x_test,y_test)*100
```
