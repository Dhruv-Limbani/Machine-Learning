```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Melbourne_housing_FULL.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Price', 'Method', 'SellerG', 'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea',
               'CouncilArea','Regionname','Propertycount']

ds = df[cols_to_use]
ds
```

```python
ds.isna().sum()
```

```python
cols_to_fill_zero = ['Distance', 'Bedroom2', 'Bathroom', 'Car', 'Propertycount']
ds[cols_to_fill_zero]=ds[cols_to_fill_zero].fillna(0)
ds
```

```python
ds.isna().sum()
```

```python
ds['Landsize']=ds['Landsize'].fillna(ds.Landsize.mean())
ds['BuildingArea']=ds['BuildingArea'].fillna(ds.BuildingArea.mean())
```

```python
ds.isna().sum()
```

```python
ds.dropna(0,'any',inplace=True)
ds
```

```python
ds.isna().sum()
```

```python
ds = pd.get_dummies(ds,drop_first=True)
ds.head()
```

```python
x = ds.drop('Price',1)
y = ds['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)         # overfitting 
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso             # l1 regulariztion
lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge             # l2 regulariztion
ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_train,y_train)
```
